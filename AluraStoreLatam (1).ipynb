### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos = {nombre: (df["Precio"]).sum() for nombre, df in tiendas.items()}

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(ingresos.keys(), ingresos.values(), color='green')
plt.title("Ingresos por tienda")
plt.ylabel("Total en COP")
plt.show()

# Mostrar tabla
for nombre, total in ingresos.items():
    print(f"{nombre}: ${total:,.0f}")

```

# 2. Ventas por categoría

```python
# Usamos Tienda 1 como ejemplo
ventas_categoria = tienda["Categoría del Producto"].value_counts()

ventas_categoria.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))
plt.title("Ventas por categoría - Tienda 1")
plt.ylabel("")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificaciones = {nombre: df["Calificación"].mean() for nombre, df in tiendas.items()}

plt.figure(figsize=(8,5))
plt.bar(calificaciones.keys(), calificaciones.values(), color='skyblue')
plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación (0 a 5)")
plt.ylim(0,5)
plt.show()

for nombre, cal in calificaciones.items():
    print(f"{nombre}: {cal:.2f}")
```

# 4. Productos más y menos vendidos

```python
def extremos_productos(df, nombre):
    productos = df["Producto"].value_counts()
    print(f"Top 5 más vendidos en {nombre}:\n{productos.head(5)}\n")
    print(f"Top 5 menos vendidos en {nombre}:\n{productos.tail(5)}\n")

# Usamos Tienda 1
extremos_productos(tienda, "Tienda 1")
```

# 5. Envío promedio por tienda

```python
envio_promedio = {nombre: df["Costo de envío"].mean() for nombre, df in tiendas.items()}

plt.figure(figsize=(8,5))
plt.bar(envio_promedio.keys(), envio_promedio.values(), color='orange')
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo en COP")
plt.show()

for nombre, envio in envio_promedio.items():
    print(f"{nombre}: ${envio:,.0f}")

```
